# 01: Visualize HWNAS Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import moocore
```

```python
df = pd.read_csv("results/all_hwnas.csv")
df_cifar10 = df[df["dataset"] == "cifar10"]
```

```python
vcs = df_cifar10["seed"].value_counts()
print(f"Total arch_idx * seed combinations: {vcs.sum()}")
vcs
```

```python
plt.scatter(df_cifar10.flop, df_cifar10.test_acc)
plt.show()
```

```python
df_cifar10["test_acc"].value_counts()
```

```python
df_nd = moocore.filter_dominated(df_cifar10[["flop", "test_acc"]], maximise = [False, True])
df_nd.sort_values("test_acc")
```

```python
plt.scatter(df_cifar10.flop, df_cifar10.test_acc)
plt.scatter(df_nd.flop, df_nd.test_acc)
plt.show()
```
